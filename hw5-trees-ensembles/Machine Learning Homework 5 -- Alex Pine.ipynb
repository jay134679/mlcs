## 6.2 $L_2$ Boosting
\begin{align}
\ell(f(x), y) & = \frac{1}{2}(f(x) - y)^2
\\
\text{MSE} & = \sum_{i=1}^{n}\ell(f(x_i), y_i)
\\
\text{MSE} & = \frac{1}{2}\sum_{i=1}^{n}\big[(f(x_i)-y_i)^2\big]
\\
\mathbf{g_m} & = \bigg(\frac{\partial(\text{MSE})}{\partial f_{m-1}(x_i)}\bigg), i = 1 ... n 
\\
\mathbf{g_m} & = \bigg( f_{m-1}(x_i) - y_i \bigg), i = 1 ... n 
\\
h_m & =\text{argmin}_{h \in \mathcal{F}} \sum_{i=1}^{n}\big[(-\mathbf{g_m}_i - h(x_i))^2\big]
\\
h_m & =\text{argmin}_{h \in \mathcal{F}} \sum_{i=1}^{n}\big[(y_i - f_{m-1}(x_i) - h(x_i))^2\big]
\end{align}

## 6.3 Binomial Boosting
\begin{align}
\ell(f(x), y) & = ln\big(1-exp(-yf(x))\big)
\\
\text{MSE} & = \sum_{i=1}^{n}\ell(f(x_i), y_i)
\\
\text{MSE} & = \sum_{i=1}^{n}\big[ln\big(1-exp(-yf(x))\big)\big]
\\
\mathbf{g_m} & = \bigg(\frac{\partial(\text{MSE})}{\partial f_{m-1}(x_i)}\bigg), i = 1 ... n 
\\
\mathbf{g_m} & = \bigg( \frac{y_i exp(-yf(x))}{1 - exp(-yf(x))} \bigg), i = 1 ... n 
\\
h_m & =\text{argmin}_{h \in \mathcal{F}} \sum_{i=1}^{n}\big[(-\mathbf{g_m}_i - h(x_i))^2\big]
\\
h_m & =\text{argmin}_{h \in \mathcal{F}} \sum_{i=1}^{n}\bigg[\big(\frac{-y_i exp(-yf(x))}{1 - exp(-yf(x))} - h(x_i)\big)^2\bigg]
